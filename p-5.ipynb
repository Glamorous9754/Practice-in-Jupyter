```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("C:/Users/ayan1/OneDrive/Documents/Course/Machine Learning Using Python/Data Folder/survey_lung_cancer.csv",header=0)
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df = df.drop(['GENDER_M'],axis=1)
```

```python
df.head()
```

```python
df_corr = df.corr()
LUNG_CANCER_corr = df_corr.LUNG_CANCER.drop(['LUNG_CANCER'],axis=0)
```

```python
LUNG_CANCER_corr
```

```python
from sklearn.linear_model import LogisticRegression

x = df[['SMOKING']]
y = df['LUNG_CANCER']

model = LogisticRegression()
model.fit(x, y)

print("Intercept: ",model.intercept_[0],"\nCoefficients: ",model.coef_[0][0])
```

```python
from sklearn.model_selection import train_test_split

x_multi = df.drop(['LUNG_CANCER'], axis=1).astype(float)
y_multi = df['LUNG_CANCER'].astype(float)  # Convert to integer

x_train, x_test, y_train, y_test = train_test_split(x_multi, y_multi, test_size=0.2, random_state=5)

model = LogisticRegression(max_iter=500)  # Increase the number of iterations
model.fit(x_train, y_train)

print("Intercept: ", model.intercept_[0])
print("Coefficients: ", model.coef_[0])
```
