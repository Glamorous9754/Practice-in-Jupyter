```python
x_multi = df.drop(['quality'],axis=1)
x_multi = x_multi.astype(float)
x_multi_c = sn.add_constant(x_multi)
```

```python
y_multi = df['quality'].astype(float)
```

```python
lm_multi_1 = sn.OLS(y_multi,x_multi_c).fit()
```

```python
lm_multi_1.summary()
```

```python
from sklearn.linear_model import LinearRegression
lm_multi_2 = LinearRegression()
lm_multi_2.fit(x_multi,y_multi)
```

```python
print(lm_multi_2.intercept_,'\n',lm_multi_2.coef_)
```
