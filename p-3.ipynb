```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/ayan1/Downloads/Dummy Data HSS.csv",header=0)
df.head()
# df.shape
```

```python
df.describe()
```

```python
df['Social Media']=df['Social Media'].fillna(df['Social Media'].mean())
df['Radio']=df['Radio'].fillna(df['Radio'].mean())
df['Sales']=df['Sales'].fillna(df['Sales'].mean())
df['TV']=df['TV'].fillna(df['TV'].mean())
df.describe()
```

```python
df.info()
```

```python
sns.jointplot(x="TV", y="Sales",data=df)
plt.show()
```

```python
sns.jointplot(x="Social Media", y="Sales",data=df)
plt.show()
```

```python
sns.jointplot(x="Radio", y="Sales",data=df)
plt.show()
```

```python
sns.countplot(x="Influencer",data=df)
plt.show()
```
