```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/ayan1/Downloads/car_price.csv",header=0,index_col=0)
df.head()
```

```python
df.info()
```

```python
df.Kilometers_Driven = df.Kilometers_Driven.fillna(df.Kilometers_Driven.mean())
df.info()
```

```python
df.Airbags = df.Airbags.fillna(df.Airbags.mean())
# df['Airbags'] = df['Airbags'].astype(object)
df.info()
```

```python
df.describe()
```

```python
sns.jointplot(x='Kilometers_Driven',y='Price',data=df)
plt.show()
```

```python
sns.violinplot(x='Kilometers_Driven',y='Price',data=df)
plt.show()
```

```python
sns.boxplot(x="Kilometers_Driven",data=df)
plt.show()
```

```python
sns.jointplot(x='Airbags',y='Price',data=df)
plt.show()
```

```python
sns.violinplot(x='Airbags',y='Price',data=df)
plt.show()
```

```python
sns.boxplot(x='Airbags',y='Price',data=df)
plt.show()
```
